```python
#import pandas library
import pandas as pd
```

```python
#load data set
df = pd.read_csv(r'C:\Users\sandu\Downloads\Data Set For Task\1) iris.csv')
print(df.head())    #show rows
print("No of rows and coloumns:",df.shape)    #show rows & coloumns count
```

```python
#check missing values of the data set

print("Missing values per column:")
print(df.isnull().sum())

print("\nTotal missing values:",df.isnull().sum().sum())
#this data set has no missing values so that count is 0
```

```python
#check duplicates count before removing duplicates
print("Duplicates before:",df.duplicated().sum())
```

```python
#remove duplicates & checck count
df.drop_duplicates(inplace = True)
print("Duplicates after:",df.duplicated().sum())
```

```python
#make standard for the species coloumn
df['species'] = df['species'].str.strip().str.title()
print(df['species'].unique())  # see all unique species names
```
