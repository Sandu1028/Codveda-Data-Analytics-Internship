```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sandu\Downloads\Data Set For Task\1) iris.csv')
print(df.head())
```

```python
plt.figure(figsize=(8,5))
df['species'].value_counts().plot(kind='bar',color=['steelblue','green','coral'],
                                   edgecolor='black')
plt.title('Number of Flowers Per Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.savefig(r'C:\Users\sandu\Downloads\Intern - Codveda Technologies\Level 1\Task3_barplot.png')
plt.show()
print("Bar Plot Saved")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['sepal_length'].values, label='Sepal Length', color='blue')
plt.plot(df['petal_length'].values, label='Petal Length', color='red')
plt.title('Sepal Length vs Petal Length Across Samples', fontsize=14)
plt.xlabel('Sample Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.savefig(r'C:\Users\sandu\Downloads\Intern - Codveda Technologies\Level 1\Task3_linechart.png')
plt.show()
print("Line chart saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, 
                x='sepal_width', 
                y='petal_width', 
                hue='species',
                palette='deep',
                s=100)
plt.title('Sepal Width vs Petal Width', fontsize=14)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.savefig(r'C:\Users\sandu\Downloads\Intern - Codveda Technologies\Level 1\Task3_scatterplot2.png')
plt.show()
print("Scatter plot saved!")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Bar plot
df['species'].value_counts().plot(kind='bar', 
                                   ax=axes[0],
                                   color=['steelblue','green','coral'],
                                   edgecolor='black')
axes[0].set_title('Species Count')
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=0)

# Line chart
axes[1].plot(df['sepal_length'].values, label='Sepal Length', color='blue')
axes[1].plot(df['petal_length'].values, label='Petal Length', color='red')
axes[1].set_title('Sepal vs Petal Length')
axes[1].set_xlabel('Sample Index')
axes[1].set_ylabel('Length (cm)')
axes[1].legend()

# Scatter plot
sns.scatterplot(data=df, 
                x='sepal_width', 
                y='petal_width', 
                hue='species',
                ax=axes[2],
                palette='deep')
axes[2].set_title('Sepal Width vs Petal Width')
axes[2].set_xlabel('Sepal Width (cm)')
axes[2].set_ylabel('Petal Width (cm)')

plt.suptitle('Iris Dataset - Basic Visualizations', fontsize=16)
plt.tight_layout()
plt.savefig(r'C:\Users\sandu\Downloads\Intern - Codveda Technologies\Level 1\all_charts.png')
plt.show()
print("Combined chart saved!")
```
